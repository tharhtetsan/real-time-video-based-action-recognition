```python
def image_stack(image_list,CHANNEL):
    width, height,_ = image_list[0].shape
    stacked_image = np.zeros((width ,height,len(image_list)*CHANNEL), np.uint8)
    
    start_index = 0
    for i,cur_img in enumerate(image_list):
        end_index = start_index + CHANNEL

        stacked_image[:,:,start_index:end_index]=cur_img
        start_index = end_index
    return stacked_image
```

```python
image_list = [img1,img2]
stacked_image = image_stack(image_list=image_list,CHANNEL = 3)
print(stacked_image.shape)
```
